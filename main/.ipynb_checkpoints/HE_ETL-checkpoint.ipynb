# Importações

# Transformações

### Limpeza Dados HE

```python
dados_HE.drop(['HE\nReceber', 'Salario Bruto', 'Justificativa'], axis = 1, inplace = True)
```

```python
dados_HE.rename(columns = {'Hora In�cio':'Hora_inicio', 
                           'Hora Fim': 'Hora_fim', 
                           'HE\nCompensar':'HE_retorno',
                           'Horas':'qtd_HE'
                          }, 
                inplace=True)
```

```python
dados_HE['HE_retorno'] = dados_HE['HE_retorno'].replace('X', 'Compensar')
dados_HE['HE_retorno'].fillna('Receber', inplace=True)
```

### Limpeza Dados Salario

```python
dados_salario.drop(['Feriado'],axis=1, inplace=True)
```

```python
dados_salario.rename(columns = {'Nome do Feriado':'Feriado', 
                           'Regime de trabalho': 'Escala_de_trabalho', 
                           'Salario Bruto': 'Salario_bruto',
                           'Salario Base': 'Salario_base',
                           'ValorHE/Hora':'Valor_Hora'
                          }, 
                inplace=True)
```

```python
dados_salario.Feriado.fillna('Nao', inplace=True)
```

```python
dados_salario.head()
```

### Transformações dos dados

```python
data_daily = dados_salario.merge(dados_HE, how='left', on='Data')
```

```python
data_daily.qtd_HE.fillna(0, inplace=True)
```

#### Criando Nova coluna condicional 

```python
#Será usado o numpy para gerar nova coluna devida a velocidade de execução

conditions = [(data_daily['Escala_de_trabalho'] == 'Folga') & (data_daily['qtd_HE'] > 0),
             (data_daily['Escala_de_trabalho'] == 'Trabalho') & (data_daily['qtd_HE'] > 0)]

choices = ['Folga', 'Jornada Estendida']


inicio = time.time()

data_daily['Tipo_HE'] = np.select(conditions, choices, default='NaN')

fim = time.time()
print(fim - inicio)
```

```python
#Seria uma outra forma para se gerar uma coluna condicional

def tipo_he(df):
    if (df['Escala_de_trabalho'] == 'Folga') & (df['qtd_HE'] > 0) :
        return 'Folga'
    elif (df['Escala_de_trabalho'] == 'Trabalho') & (df['qtd_HE'] > 0) :
        return 'Jornada Estendida'
    else:
        return 'NaN'
    
inicio = time.time()

data_daily['Tipo_HE2'] = data_daily.apply(tipo_he, axis=1)

fim = time.time()
print(fim - inicio)
```

```python
filtro1=data_daily.qtd_HE>0
filtro2=data_daily.Escala_de_trabalho == 'Folga'

data_daily.Tipo_HE = data_daily.loc[filtro1 & filtro2].Tipo_HE.replace('NaN', 'Folga')
```

```python
filtro1=data_daily.qtd_HE>0
filtro2=data_daily.Escala_de_trabalho == 'Trabalho'

data_daily.Tipo_HE = data_daily.Tipo_HE.loc[filtro1 & filtro2].replace('NaN','Jornada Estendida')
```
