## Limpeza dos Dados

```python
#Função para exibir a quantidade absoluta e relativa de dados ausentes de cada coluna de um dataframe df
def data_quality(df):
    qtd_rows = df.shape[0]
    data_missing = df.isnull().sum()
    data_missing_percent = {}
    for col_name in df.columns:
        data_missing_percent[col_name] = [data_missing[col_name], str(round(100*data_missing[col_name]/qtd_rows, 3))+' %']
    
    return data_missing_percent
        
```

## Transformações dos dados

### Transformando a Tabela salario

```python
data_alt_func = dados_Escala.loc[dados_Escala['Cargo'] == 'COD'].Data.values[0].astype('datetime64[D]')
```

```python
print(data_alt_func)
```

```python
#Cria a coluna 'Salario Bruto' no dataset dados_Salario
#Para datas menores do que a de alteração de função o salário base é multiplicado por 1.3
#Para datas maiores o salario base é mulltiplicado por 1.35
dados_Salario['Salario Bruto']= dados_Salario.apply(lambda x: 
                                                    x['Salario Base'] * 1.30 if x['Data inicial'] < data_alt_func
                                                   else x['Salario Base'] * 1.35,
                                                   axis=1)
```

```python
#Cria a coluna 'Valor_HE_por_hora' no dataset dados_Salario
#Para datas menores do que a de alteração de função o salário bruto é dividido por 200h
#Para datas maiores o salario base é salário bruto é dividido por 180h
dados_Salario['Valor_HE_por_hora']= dados_Salario.apply(lambda x: 
                                                    x['Salario Bruto'] / 200 if x['Data inicial'] < data_alt_func
                                                   else x['Salario Bruto'] /180,
                                                   axis=1)
```

### Transformando a Tabela de escala de trabalho em um diário de trabalho

```python
#definição da localidade para retorno em portugues das funcoes 'dt.day_name' e 'dt.month_name'
import locale
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

### Criando Nova coluna condicional no dataset diario_de_trabalho

```python
#Será usado o numpy para gerar nova coluna devida a velocidade de execução

#condições
conditions = [(diario_de_trabalho['Regime de trabalho'] == 'Folga') & (diario_de_trabalho['qtd_HE'] > 0),
             (diario_de_trabalho['Regime de trabalho'] == 'Trabalho') & (diario_de_trabalho['qtd_HE'] > 0)]

#escolhas
choices = ['Folga', 'Jornada Estendida']

#Criação da coluna 'Tipo_HE' no dataset diario_de_trabalho
diario_de_trabalho['Tipo_HE'] = np.select(conditions, choices, default=np.NaN)
```

```python
#lista com a ordem desejadas das colunas do dataset diario_de_trabalho
columns = ['Data', 'Dia da semana', 'Mes', 'Ano', 'Feriado', 'Cargo', 'Regime de trabalho', 
           'Salario Base', 'Salario Bruto', 'Valor_HE_por_hora',
           'Hora_inicio', 'Hora_fim', 'qtd_HE', 'HE_retorno', 'Tipo_HE']

#reordenação das colunas conforme lista 'columns' estabelecida anteriormente
diario_de_trabalho = diario_de_trabalho.reindex(columns=columns)
```
